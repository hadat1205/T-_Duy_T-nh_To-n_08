```python

def fib(n):
   if n<=0:
    return False
   if n==1:
    return 1
   else:
    a=0
    b=1
    for i in range(2,n+1):
      c=a+b
      a=b
      b=c
    return b
def test_first_two():
    assert fib(0) == 0, "fib(0) should be 0"
    assert fib(1) == 1, "fib(1) should be 1"

def test_small_numbers():
    assert fib(2) == 1, "fib(2) should be 1"
    assert fib(3) == 2, "fib(3) should be 2"
    assert fib(4) == 3, "fib(4) should be 3"

def test_large_number():
    assert fib(10) == 55, "fib(10) should be 55"
if __name__ == "__main__":
  test_small_numbers()
  test_first_two()
  print("chay thanh cong")
```

```python
#W5a7
def calculator(a, b, op):
    if op == '+':
        return ('%.2f' % (a + b))
    elif op == '-':
        return ('%.2f' % (a - b))
    elif op == '*':
        return ('%.2f' % (a * b))
    elif op == '/':
        return ('%.2f' % (a / b))
 
a, op, b = input().split()
print(calculator(float(a), float(b), op))
```

```python
#w5a8
def khoangcach(x, y):
    b = x ^ y
    distance = 0
    
    while b > 0:
        b = b & (b - 1)
        distance += 1
        
    return distance
a,b=map(int, input().split())
khoangcach(a,b)
```

```python
#w5a10
def khoangcachhamming(a,b):
    map_a_to_b={}
    map_b_to_a={}
    for i in range(0,len(a)):
        chara=a[i]
        charb=b[i]
        if chara in map_a_to_b:
           if  map_a_to_b[chara]!=charb:
               return False
        else:
            map_a_to_b[chara]=charb
        if charb in map_b_to_a:
            if map_b_to_a[charb]!=chara:
             return False
        else:
            map_b_to_a[charb]=chara
    return True
a,b=input().split()
khoangcachhamming(a,b)
```
