```python
#w8a1
class Cylinder:
    pi=3.14
    def __init__(seft,r,h):
        seft.r=r
        seft.h=h
    def dientich(seft):
        return 2*seft.pi*seft.r*seft.h
    def thetich(seft):
        return seft.pi*seft.h*(seft.r**2)
       
r,h=map(float,input().split())
cylinder=Cylinder(r,h)
print(f"{cylinder.dientich():.2f} {cylinder.thetich():.2f}")
```

```python
#w8a2
class Date:
    def __init__(self, ngay, thang, nam):
        self.ngay = ngay
        self.thang = thang
        self.nam = nam


    def ktra(self):
        if self.nam<0 or self.thang<1 or self.ngay<1 or self.thang>12 or self.ngay >31:
            return False
        else:
            ngaytrongthang=[31,28,31,30,31,30,31,31,30,31,30,31]
            if self.nam%400==0 or (self.nam%4==0 and self.nam%100!=0):
                ngaytrongthang[1]=29
            return self.ngay<=ngaytrongthang[self.thang-1]
        

    def ngayke(self):
        ngaytrongthang=[31,28,31,30,31,30,31,31,30,31,30,31]
        if self.nam%400==0 or (self.nam%4==0 and self.nam%100!=0):
            ngaytrongthang[1]=29
        day=self.ngay+1
        month=self.thang
        year=self.nam

        if day>ngaytrongthang[self.thang-1]:
            day=1
            month+=1
            if month>12:
                month=1
                year+=1
        return f"{day:02d} {month:02d} {year:04d}"
    

ngay,thang,nam=map(int,input().split())
date=Date(ngay,thang,nam)
if not date.ktra():
    print("INVALID")
else:
    print(date.ngayke())
         
            
    
```

```python
#w8a4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            return "Không thể chia cho 0"
        return self.a / self.b

    def power(self):
        return self.a ** self.b

    def mod(self):
        if self.b == 0:
            return "Không thể chia cho 0"
        return self.a % self.b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b


# ===== Chương trình chính =====
a = float(input("Nhập số thứ nhất: "))
b = float(input("Nhập số thứ hai: "))

calc = Calculator(a, b)

while True:
    print("\n===== MENU =====")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Lũy thừa")
    print("6. Chia lấy dư")
    print("7. Nhập 2 số mới")

    choice = int(input("Chọn phép toán (1-7): "))

    if choice == 1:
        print("Kết quả:", calc.add())
    elif choice == 2:
        print("Kết quả:", calc.subtract())
    elif choice == 3:
        print("Kết quả:", calc.product())
    elif choice == 4:
        print("Kết quả:", calc.divide())
    elif choice == 5:
        print("Kết quả:", calc.power())
    elif choice == 6:
        print("Kết quả:", calc.mod())
    elif choice == 7:
        a = float(input("Nhập số thứ nhất mới: "))
        b = float(input("Nhập số thứ hai mới: "))
        calc.set_numbers(a, b)
        print("Đã cập nhật 2 số mới!")
    else:
        print("Lựa chọn không hợp lệ!")

    thoat = input("Bạn có muốn thoát? ")
    if thoat.lower() == "yes":
        break
```

```python
#w8a5
class ShoppingCart:
    def __init__(self):
        self.cart = []

    def add_product(self, name, price):
        self.cart.append({"name": name, "price": price})

    def remove_product(self, name):
        for item in self.cart:
            if item["name"] == name:
                self.cart.remove(item)
                return "Đã xóa sản phẩm"
        return "Không tìm thấy sản phẩm"

    def is_empty(self):
        return len(self.cart) == 0

    def total_price(self):
        total = 0
        for item in self.cart:
            total += item["price"]
        return total

    def show_products(self):
        if self.is_empty():
            print("Giỏ hàng rỗng")
        else:
            for item in self.cart:
                print(item["name"], "-", item["price"])

    def clear_cart(self):
        self.cart.clear()


# ===== Chương trình chính =====
cart = ShoppingCart()

while True:
    print("\n===== MENU =====")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = int(input("Chọn chức năng (2-7): "))

    if choice == 2:
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá: "))
        cart.add_product(name, price)

    elif choice == 3:
        name = input("Nhập tên sản phẩm cần xóa: ")
        print(cart.remove_product(name))

    elif choice == 4:
        print("Giỏ hàng rỗng:", cart.is_empty())

    elif choice == 5:
        print("Tổng tiền:", cart.total_price())

    elif choice == 6:
        cart.show_products()

    elif choice == 7:
        cart.clear_cart()
        print("Đã xóa toàn bộ giỏ hàng")

    else:
        print("Lựa chọn không hợp lệ")

    exit_program = input("Có muốn thoát khỏi chương trình không? ")
    if exit_program.lower() == "yes":
        break
```

```python
#ww8a6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k


# ===== Chương trình chính =====
w, h, k = map(float, input().split())

rect = Rectangle(w, h)
rect.scale(k)

print(f"{rect.area():.2f} {rect.perimeter():.2f}")
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def withdraw(self, amount):
        # Nếu rút quá số dư thì bỏ qua giao dịch
        if 0 < amount <= self.balance:
            self.balance -= amount

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

# Phần chạy chương trình
try:
    # Đọc dòng đầu tiên: owner và balance
    first_line = input().split()
    owner = first_line[0]
    initial_balance = float(first_line[1])
    
    account = BankAccount(owner, initial_balance)
    
    # Đọc số lượng lệnh
    n = int(input())
    
    for _ in range(n):
        line = input().split()
        cmd = line[0]
        amount = float(line[1])
        
        if cmd == 'DEPOSIT':
            account.deposit(amount)
        elif cmd == 'WITHDRAW':
            account.withdraw(amount)
            
    print(f"{account.balance:.2f}")

except Exception as e:
    print("Lỗi nhập liệu:", e)
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8.0:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5.0:
            return "Average"
        else:
            return "Poor"

# Phần chạy chương trình
try:
    name = input().strip()
    n = int(input())
    
    student = Student(name)
    
    for _ in range(n):
        # Input dạng: Subject Score (ví dụ: Math 9)
        parts = input().split()
        # Điểm là phần tử cuối cùng
        score = float(parts[-1]) 
        # Tên môn học là phần còn lại (không dùng trong tính toán nhưng cần tách ra)
        subject = " ".join(parts[:-1])
        
        student.add_score(subject, score)
        
    average_score = student.avg()
    ranking = student.rank()
    
    print(f"{student.name} {average_score:.2f} {ranking}")

except Exception as e:
    print("Lỗi:", e)
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, author_name):
        count = 0
        for book in self.books:
            # So sánh chính xác tên tác giả
            if book.author == author_name:
                count += 1
        return count

    def find_by_year(self, year):
        count = 0
        for book in self.books:
            if book.year == year:
                count += 1
        return count

# Phần chạy chương trình
try:
    n = int(input())
    lib = Library()
    
    for _ in range(n):
        line = input().strip()
        
        if line.startswith("ADD"):
            # Cấu trúc: ADD title; author; year
            # Cắt bỏ chữ "ADD " đầu tiên (4 ký tự)
            content = line[4:] 
            parts = content.split(';')
            title = parts[0].strip()
            author = parts[1].strip()
            year = int(parts[2].strip())
            lib.add(Book(title, author, year))
            
        elif line.startswith("COUNTYEAR"):
            # Cấu trúc: COUNTYEAR year
            year_query = int(line.split()[1])
            print(lib.find_by_year(year_query))
            
        elif line.startswith("COUNT"):
            # Cấu trúc: COUNT author
            # Tên tác giả có thể có dấu cách, lấy phần sau chữ "COUNT "
            author_query = line[6:].strip()
            print(lib.count_by_author(author_query))

except Exception as e:
    print("Lỗi:", e)
```

```python
import math

class Point3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

def subtract(p1, p2):
    return Point3D(p1.x - p2.x, p1.y - p2.y, p1.z - p2.z)

def cross_product(v1, v2):
    return Point3D(
        v1.y * v2.z - v1.z * v2.y,
        v1.z * v2.x - v1.x * v2.z,
        v1.x * v2.y - v1.y * v2.x
    )

def dot_product(v1, v2):
    return v1.x * v2.x + v1.y * v2.y + v1.z * v2.z

def magnitude(v):
    return math.sqrt(v.x**2 + v.y**2 + v.z**2)

try:
    points = []
    # Đọc 4 điểm A, B, C, D
    for _ in range(4):
        coords = list(map(float, input().split()))
        points.append(Point3D(coords[0], coords[1], coords[2]))
    
    A, B, C, D = points[0], points[1], points[2], points[3]
    
    # Mặt phẳng (A, B, C): Vectơ pháp tuyến n1 = AB x AC
    AB = subtract(B, A)
    AC = subtract(C, A)
    n1 = cross_product(AB, AC)
    
    # Mặt phẳng (B, C, D): Vectơ pháp tuyến n2 = BC x BD
    BC = subtract(C, B)
    BD = subtract(D, B)
    n2 = cross_product(BC, BD)
    
    # Tính cos góc
    dot = dot_product(n1, n2)
    mag1 = magnitude(n1)
    mag2 = magnitude(n2)
    
    # Tránh chia cho 0
    if mag1 == 0 or mag2 == 0:
        print("0.00")
    else:
        cos_phi = abs(dot) / (mag1 * mag2)
        # Đảm bảo giá trị trong [-1, 1] do sai số float
        cos_phi = max(min(cos_phi, 1), -1)
        
        angle_rad = math.acos(cos_phi)
        angle_deg = math.degrees(angle_rad)
        
        print(f"{angle_deg:.2f}")

except Exception as e:
    print(e)
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def multiply(self, other):
        new_real = self.real * other.real - self.imag * other.imag
        new_imag = self.real * other.imag + self.imag * other.real
        return ComplexNumber(new_real, new_imag)

try:
    # Nhập số phức 1
    line1 = list(map(float, input().split()))
    c1 = ComplexNumber(line1[0], line1[1])
    
    # Nhập số phức 2
    line2 = list(map(float, input().split()))
    c2 = ComplexNumber(line2[0], line2[1])
    
    result = c1.multiply(c2)
    
    # In ra: phần thực phần ảo
    # Sử dụng format g để xóa số 0 thừa nếu cần, hoặc .2f tùy đề
    # Theo test case mẫu: -13 82 -> in số nguyên nếu tròn, số thực nếu lẻ
    # Tuy nhiên để đơn giản và khớp với đa số bài chấm tự động float:
    print(f"{int(result.real) if result.real.is_integer() else result.real} {int(result.imag) if result.imag.is_integer() else result.imag}")

except:
    print("Lỗi nhập liệu")
```

```python
import math

class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3

    def get_area(self):
        a, b, c = self.e1, self.e2, self.e3
        # Kiểm tra điều kiện tam giác: tổng 2 cạnh > cạnh còn lại
        if (a + b > c) and (a + c > b) and (b + c > a):
            p = (a + b + c) / 2
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        else:
            return -1

try:
    # Nhập 3 cạnh trên 1 dòng
    edges = list(map(int, input().split()))
    t = Triangle(edges[0], edges[1], edges[2])
    
    area = t.get_area()
    
    if area == -1:
        print("invalid")
    else:
        print(f"{area:.2f}")

except:
    print("invalid")
```
