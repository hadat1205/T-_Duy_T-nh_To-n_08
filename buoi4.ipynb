```python

def sodao(n):
    m = 0
    while n != 0:
        b = n % 10
        m = m * 10 + b
        n //= 10
    return m

n=int(input("Nhập số nguyên dương n: "))
dem=0
while n!=sodao(n):
    n+=sodao(n)
    dem+=1
print(dem,n)
```

```python
#w4a32
def kytudacbiet(x):
    return not x.isalnum()

s = input()

hoa = False
thuong = False
so = False
dacbiet = False

for c in s:
    if c.isupper():
        hoa = True
    elif c.islower():
        thuong = True
    elif c.isdigit():
        so = True
    elif kytudacbiet(c):
        dacbiet = True

if hoa and thuong and so and dacbiet:
    print("mk manh")
else:
    print("mk yeu")
```
