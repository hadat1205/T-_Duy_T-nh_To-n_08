```python
#w6a10
a = list(map(int, input().split()))
k = int(input())

c = []

for i in range(len(a)):
    for j in range(i + 1, len(a)):   
        if a[i] + a[j] == k :
            d=(a[i],a[j])
            if d not in c:
                c.append(d)
c.sort()

print(c)
```

```python
#w6a17
n = int(input())

a = []
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)

# In đường chéo chính
print("Duong cheo chinh:")
for i in range(n):
    print(a[i][i], end=" ")
print()

# In đường chéo phụ
print("Duong cheo phu:")
for i in range(n):
    print(a[i][n - 1 - i], end=" ")
```
