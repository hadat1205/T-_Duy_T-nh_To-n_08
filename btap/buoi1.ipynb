```python
#bai1
def sochuso(n):
    chuso=0
    n=abs(n)
    while n>0:
        chuso+=1
        n//=10
    return chuso
n=int(input())
sochuso(n)
```

```python
#bai2
def sochan(a):
    sochan=0
    for i in range(0,len(a)):
        if int(a[i])%2==0:
            sochan+=1
    return sochan
a=input(list()).split()
sochan(a)
```
