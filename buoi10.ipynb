```python
#w10a3
def is_happy_number(n: int) -> bool:
    seen = set() 
    while n != 1 and n not in seen:
        seen.add(n)
        s = 0
        while n > 0:
            digit = n % 10
            s += digit * digit
            n //= 10
        n = s
    return n == 1
try:
    a = int(input("Nhap so nguyen duong a: "))
    if a <= 0:
        print("a khong phai so nguyen duong")
    else:
        if is_happy_number(a):
            print("a la so hanh phuc")
        else:
            print("a khong phai so hanh phuc")

except ValueError:
    print("Gia tri nhap vao khong hop le (khong phai so nguyen)")
```
