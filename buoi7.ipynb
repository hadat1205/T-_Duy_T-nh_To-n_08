```python
#W7a3
def bubble_sort(a):
    n = len(a)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a
a=list(map(int,input().split()))
print(bubble_sort(a))
```

```python
def demsolanxuathien(x, arr):
    return arr.count(x)

arr = list(map(int, input().split()))

max_count = 0      # số lần xuất hiện lớn nhất
value = 0    # giá trị xuất hiện nhiều nhất

for i in range(len(arr)):
    cnt = demsolanxuathien(arr[i], arr)
    if cnt > max_count:
        max_count = cnt
        value = arr[i]

print(f"{value} xuất hiện nhiều nhất, sớm nhất, số lần xuất hiện {max_count}")
```

```python
def solve_w7a8():
 
    try:
        # Nhập intervals (dạng list of lists)
        intervals_str = input("Nhập intervals (VD: [[1, 4], [2, 4]]): ")
        intervals = eval(intervals_str) 
        
        # Nhập queries (dạng list)
        queries_str = input("Nhập queries (VD: [2, 3, 4, 5]): ")
        queries = eval(queries_str)

        results = []
        
        for q in queries:
            min_len = float('inf')
            found = False
            
            for start, end in intervals:
                if start <= q <= end:
                    curr_len = end - start + 1
                    if curr_len < min_len:
                        min_len = curr_len
                        found = True
            
            if found:
                results.append(min_len)
            else:
                results.append(-1)
                
        # In kết quả dạng list
        print(results)
        
    except:
        print("Lỗi nhập liệu")

if __name__ == "__main__":
    solve_w7a8()
```

```python
import bisect

def solve_w7a9():
    try:
        # Nhập list nums
        nums_str = input("Nhập nums (VD: [5, 2, 6, 1]): ")
        nums = eval(nums_str)
        
        result = []
        sorted_seen = []
        
        # Duyệt ngược từ phải sang trái
        for num in reversed(nums):
            # Tìm vị trí của num trong danh sách đã sắp xếp các số bên phải
            # bisect_left trả về vị trí mà mọi phần tử bên trái đều nhỏ hơn num
            count = bisect.bisect_left(sorted_seen, num)
            result.append(count)
            
            # Thêm num vào danh sách sorted_seen để dùng cho vòng lặp sau
            bisect.insort(sorted_seen, num)
            
        # Đảo ngược lại kết quả để đúng thứ tự ban đầu
        print(result[::-1])
        
    except:
        print("Lỗi nhập liệu")

if __name__ == "__main__":
    solve_w7a9()
```

```python
import bisect

def solve_w7a10():
    try:
        # Nhập nums và m
        nums_str = input("Nhập nums (VD: [3, 3, 9, 9, 5]): ")
        nums = eval(nums_str)
        m = int(input("Nhập m: "))
        
        max_mod_sum = 0
        prefix_sum = 0
        # Danh sách lưu các prefix_sum % m đã gặp
        sorted_prefixes = [0] 
        
        for num in nums:
            prefix_sum = (prefix_sum + num) % m
            
            # Trường hợp 1: Lấy chính prefix_sum hiện tại
            max_mod_sum = max(max_mod_sum, prefix_sum)
            
            # Trường hợp 2: Tìm phần tử nhỏ nhất trong sorted_prefixes lớn hơn prefix_sum hiện tại
            # index này là vị trí của phần tử >= prefix_sum + 1
            idx = bisect.bisect_left(sorted_prefixes, prefix_sum + 1)
            
            if idx < len(sorted_prefixes):
                # Nếu tồn tại, ta lấy hiệu
                curr_max = (prefix_sum - sorted_prefixes[idx] + m) % m
                max_mod_sum = max(max_mod_sum, curr_max)
            
            bisect.insort(sorted_prefixes, prefix_sum)
            
        print(max_mod_sum)

    except:
        print("Lỗi nhập liệu")

if __name__ == "__main__":
    solve_w7a10()
```

```python
import re

def solve_w7a11():
    text = input("Nhập đoạn văn bản: ")
    target = input("Nhập từ cần tìm: ").strip()
    
    # Sử dụng Regular Expression để tách từ, loại bỏ dấu câu
    # \b là ranh giới từ (word boundary)
    # re.IGNORECASE để không phân biệt hoa thường
    pattern = r'\b' + re.escape(target) + r'\b'
    matches = re.findall(pattern, text, re.IGNORECASE)
    
    print(len(matches))

if __name__ == "__main__":
    solve_w7a11()
```

```python
def solve_w7a12():
    options = []
    # Đọc 4 dòng input
    print("Nhập 4 phương án (mỗi phương án 1 dòng):")
    for _ in range(4):
        options.append(input().strip())
        
    # Sắp xếp theo thứ tự từ điển
    sorted_options = sorted(options)
    
    # In ra cách nhau bởi dấu cách
    print(" ".join(sorted_options))

if __name__ == "__main__":
    solve_w7a12()
```

```python
def solve_w7a14():
    try:
        n = int(input("Nhập n: "))
        # Nhập n số nguyên trên dòng tiếp theo
        arr = list(map(int, input("Nhập dãy số: ").split()))
        
        # Tìm các index có giá trị 7
        indices = [str(i) for i, x in enumerate(arr) if x == 7]
        
        if not indices:
            print("Not found")
        else:
            # Sắp xếp giảm dần (theo yêu cầu "vị trí lớn hơn đứng trước")
            indices.sort(key=lambda x: int(x), reverse=True)
            print(" ".join(indices))
            
    except:
        print("Not found")

if __name__ == "__main__":
    solve_w7a14()
```

```python
def solve_w7a15():
    try:
        n = int(input("Nhập số lượng khách n: "))
        guests = []
        for _ in range(n):
            guests.append(input().strip())
            
        if "Nemo" in guests:
            idx = guests.index("Nemo")
            
            # Tìm người đứng trước (vòng tròn nên dùng modulo)
            # Python xử lý số âm trong index list tốt (-1 là cuối), 
            # nhưng để rõ ràng logic toán học: (idx - 1 + n) % n
            prev_guest = guests[(idx - 1 + n) % n]
            
            # Tìm người đứng sau
            next_guest = guests[(idx + 1) % n]
            
            print(f"{prev_guest} and {next_guest}")
        else:
            print("Nemo không có trong danh sách")
            
    except Exception as e:
        print("Lỗi:", e)

if __name__ == "__main__":
    solve_w7a15()
```
